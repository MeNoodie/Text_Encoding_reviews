```python
import pandas as pd
import string

# 1. Setup
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # Convert to string to handle any NaN values
    Review_text = str(text).lower() 
    
    # Remove punctuation
    Review_text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Remove stopwords and rejoin into a string
    words = Review_text.split()
    cleaned_Review = [w for w in words if w not in stop_words]
    
    return " ".join(cleaned_Review)

# 3. Apply to your column
df['cleaned_Title'] = df['Review Title'].apply(clean_text)
df["Cleaned_review"] = df["Review Description"].apply(clean_text)

# 4. Convert to list (to avoid the inhomogeneous shape error)
title = df['cleaned_Title'].values.tolist()
review = df["Cleaned_review"].values.tolist()
```

```python

import numpy as np
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
# Use cleaned titles already created in notebook
titles = df['cleaned_Title'].fillna('').astype(str)
tokenized_titles = [t.split() for t in titles]

# 1) Proper word-level OHE
mlb_word = MultiLabelBinarizer()
ohe_word = mlb_word.fit_transform(tokenized_titles)

# 2) BoW and TF-IDF
bow_vec = CountVectorizer()
bow_mat = bow_vec.fit_transform(titles)

tfidf_vec = TfidfVectorizer()
tfidf_mat = tfidf_vec.fit_transform(titles)
# Helper: density and zero percentage
def matrix_stats(name, mat):
    total = mat.shape[0] * mat.shape[1]
    non_zero = np.count_nonzero(mat) if isinstance(mat, np.ndarray) else mat.nnz
    density = non_zero / total
    zero_pct = (1 - density) * 100
    return {
        'Method': name,
        'Shape': f'{mat.shape[0]} x {mat.shape[1]}',
        'NonZeroDensity': round(density, 6),
        'ZeroPercent': round(zero_pct, 2)
    }

comparison_df = pd.DataFrame([
    matrix_stats('OHE (Word-Level)', ohe_word),
    matrix_stats('BoW', bow_mat),
    matrix_stats('TF-IDF', tfidf_mat)
])
print('Task 4 + 5 Comparison Table:')
display(comparison_df)

# 3) TF-IDF importance: high-IDF words vs common words
idf_df = pd.DataFrame({
    'word': tfidf_vec.get_feature_names_out(),
    'idf': tfidf_vec.idf_
}).sort_values('idf', ascending=False)

print('Top 15 High-IDF (more unique/informative) words:')
display(idf_df.head(15))

freq_series = pd.Series(Title_words).sort_values(ascending=False)
common_words = set(freq_series.head(15).index)
common_idf = idf_df[idf_df['word'].isin(common_words)].sort_values('idf')
print('Common frequent words and their IDF (usually lower weight):')
display(common_idf.head(15))

# 4) Task 6 concise answers (kept in notebook output, no Task 7)
answers = {
    'Why BoW fails semantically': 'Ignores context, word order, and synonym relationships.',
    'When to use BoW': 'Simple baseline, small/controlled datasets, fast experiments.',
    'When to use TF-IDF': 'Need better discrimination by down-weighting very common words.',
    'TF-IDF limitations': 'No deep semantics/context, sensitive to noisy rare terms, order ignored.'
}
print('Task 6 Answers:')
for k, v in answers.items():
    print(f'- {k}: {v}')
```
